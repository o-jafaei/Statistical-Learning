```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn import linear_model
from sklearn.metrics import r2_score
```

```python
M=10
x_train=np.linspace(0,1,10)
eps_train=np.random.normal(loc=0.0, scale=0.3, size=M)
t_train=np.sin(4*np.pi*x_train)+eps_train
```

```python
N=100
x_valid=np.linspace(0,1,100)
eps_valid=np.random.normal(loc=0.0, scale=0.3, size=N)
t_valid=np.sin(4*np.pi*x_valid)+eps_valid
```

```python
poly0=PolynomialFeatures(degree=0)
x_train0=poly0.fit_transform(x_train.reshape(-1, 1))

poly1=PolynomialFeatures(degree=1)
x_train1=poly1.fit_transform(x_train.reshape(-1, 1))

poly2=PolynomialFeatures(degree=2)
x_train2=poly2.fit_transform(x_train.reshape(-1, 1))

poly3=PolynomialFeatures(degree=3)
x_train3=poly3.fit_transform(x_train.reshape(-1, 1))

poly4=PolynomialFeatures(degree=4)
x_train4=poly4.fit_transform(x_train.reshape(-1, 1))

poly5=PolynomialFeatures(degree=5)
x_train5=poly5.fit_transform(x_train.reshape(-1, 1))

poly6=PolynomialFeatures(degree=6)
x_train6=poly6.fit_transform(x_train.reshape(-1, 1))

poly7=PolynomialFeatures(degree=7)
x_train7=poly7.fit_transform(x_train.reshape(-1, 1))

poly8=PolynomialFeatures(degree=8)
x_train8=poly8.fit_transform(x_train.reshape(-1, 1))

poly9=PolynomialFeatures(degree=9)
x_train9=poly9.fit_transform(x_train.reshape(-1, 1))
```

# Part C.

```python
from sklearn.linear_model import Ridge
r2=[]
MSE_valid_r=[]
MSE_train_r=[]
iterations=[]
c=[]
r_coefs=[]
for i in range(1,11):
    ridge=Ridge(alpha=10**(-i))
    iterations.append(10**(-i))
    ridge.fit(x_train9,t_train)
    t_hat_v_ridge=ridge.predict(x_valid9)
    t_hat_t_ridge=ridge.predict(x_train9)
    MSE_valid_r.append(np.mean((t_valid-t_hat_v_ridge)**2))
    MSE_train_r.append(np.mean((t_train-t_hat_t_ridge)**2))
    r2.append(r2_score(t_valid,t_hat_v_ridge))
    r_coefs.append(ridge.coef_[1:10])
```

```python
plt.plot(iterations,r2,'o',label='R**2 Score')
plt.plot(iterations,MSE_valid_r,'*',label='MSE')
plt.xlabel('Value of alpha')
plt.legend(loc='upper right')
plt.xscale('log')
plt.title('R**2 and MSE(Validation) of Ridge')
```

```python
from sklearn.linear_model import Lasso
r2=[]
MSE_valid_l=[]
MSE_train_l=[]
l_coefs=[]
for i in range(1,11):
    lasso=Lasso(alpha=10**(-i))
    lasso.fit(x_train9,t_train)
    t_hat_v_lasso=lasso.predict(x_valid9)
    t_hat_t_lasso=lasso.predict(x_train9)
    MSE_valid_l.append(np.mean((t_valid-t_hat_v_lasso)**2))
    MSE_train_l.append(np.mean((t_train-t_hat_t_lasso)**2))
    r2.append(r2_score(t_valid,t_hat_v_lasso))
    l_coefs.append(lasso.coef_[1:10])
```

```python
plt.plot(iterations,r2,'o',label='R**2 Score')
plt.plot(iterations,MSE_valid_l,'*',label='MSE')
plt.xlabel('Value of alpha')
plt.legend(loc='upper left')
plt.xscale('log')
plt.title('R**2 and MSE(Validation) of Lasse')
```
