```python
import numpy as np
import matplotlib.pyplot as plt
```

# Part A.

```python
M=10
x_train=np.random.uniform(low=0.0, high=1.0, size=M)
x_train=np.sort(x_train)
eps_train=np.random.normal(loc=0.0, scale=0.09, size=M)
t_train=np.sin(4*np.pi*x_train)+eps_train
```

```python
N=100
x_valid=np.random.uniform(low=0.0, high=1.0, size=N)
x_valid=np.sort(x_valid)
eps_valid=np.random.normal(loc=0.0, scale=0.09, size=N)
t_valid=np.sin(4*np.pi*x_valid)+eps_valid
```

# Part B.

```python
f0=np.polyfit(x_train,t_train,0)
f1=np.polyfit(x_train,t_train,1)
f2=np.polyfit(x_train,t_train,2)
f3=np.polyfit(x_train,t_train,3)
f4=np.polyfit(x_train,t_train,4)
f5=np.polyfit(x_train,t_train,5)
f6=np.polyfit(x_train,t_train,6)
f7=np.polyfit(x_train,t_train,7)
f8=np.polyfit(x_train,t_train,8)
f9=np.polyfit(x_train,t_train,9)
```

```python
y_hat_0=np.poly1d(f0)
y_hat_1=np.poly1d(f1)
y_hat_2=np.poly1d(f2)
y_hat_3=np.poly1d(f3)
y_hat_4=np.poly1d(f4)
y_hat_5=np.poly1d(f5)
y_hat_6=np.poly1d(f6)
y_hat_7=np.poly1d(f7)
y_hat_8=np.poly1d(f8)
y_hat_9=np.poly1d(f9)
```
